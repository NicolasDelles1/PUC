```python
import skimage as sk # load skimage libraries
from skimage import io
I = sk.io.imread('vangogh.jpg') # read image using imread function
```

```python
IG = sk.color.rgb2gray(I) #transition from color to gris
IG.shape
IG=IG*255
```

```python
fig = plt.figure(figsize=(10,6))
sk.io.imshow(IG) # visualize image using imshow function
sk.io.show()
```
